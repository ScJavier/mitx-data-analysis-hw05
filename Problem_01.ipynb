# Problema 1. a)

```python
n_rows = 504
n_cols = 555

column_indexes = [i for i in range(0, 555)]
file_indexes = [i+1 for i in range(0, 100)]

sum_vel = pd.DataFrame(np.zeros([n_rows, n_cols]), columns=column_indexes)
sum_vel_sqrd = pd.DataFrame(np.zeros([n_rows, n_cols]), columns=column_indexes)
```

```python
for file_index in file_indexes: 
    file_name_u = './release_gp_homework_data/OceanFlow/{}u.csv'.format(file_index)
    file_name_v = './release_gp_homework_data/OceanFlow/{}v.csv'.format(file_index)
    
    df_temp_u = pd.read_csv(file_name_u, names=column_indexes)
    df_temp_v = pd.read_csv(file_name_v, names=column_indexes)
    
    df_vel = np.sqrt(df_temp_u.loc[:,:]**2 + df_temp_v.loc[:,:]**2)
    
    sum_vel.loc[:,:] = sum_vel.loc[:,:] + df_vel.loc[:,:]
    sum_vel_sqrd.loc[:,:] = sum_vel_sqrd.loc[:,:] + df_vel.loc[:,:]**2
```

```python
df_mask = pd.read_csv('./release_gp_homework_data/OceanFlow/mask.csv', names = column_indexes)
```

```python
df_vel_mean = sum_vel.loc[:,:] / 100
df_vel_var = (1/100)*(sum_vel_sqrd.loc[:,:] - 100*df_vel_mean.loc[:,:]**2)
```

```python
df_vel_var[df_vel_var.loc[:,:] == 0] = np.nan
```

```python
min_y = np.argmin(df_vel_var.min(axis=1)) # mínimo sobre las filas
min_x = np.argmin(df_vel_var.min(axis=0)) # mínimo sobre las columnas
```

# Problema 1. b)

```python
prev_max = -np.inf
file_max = 0
x_max_velx = 0
y_max_velx = 0

for file_index in file_indexes: 
    file_name_u = './release_gp_homework_data/OceanFlow/{}u.csv'.format(file_index)
    file_name_v = './release_gp_homework_data/OceanFlow/{}v.csv'.format(file_index)
    
    df_temp_u = pd.read_csv(file_name_u, names=column_indexes)
    
    curr_max = df_temp_u.max().max()
    
    max_y = 3*np.argmax(df_temp_u.max(axis=1)) # mínimo sobre las filas
    max_x = 3*np.argmax(df_temp_u.max(axis=0)) # mínimo sobre las columnas
    
    if curr_max >= prev_max:
        prev_max = curr_max
        x_max_velx = max_x
        y_max_velx = max_y
        file_max = file_index
    
    print('Iteración: ', file_index)
    print('max: ', prev_max)
    print('file_max: ', file_max)
    print('hour_max: ', 3*(file_max - 1))
    print('x (km): ', x_max_velx)
    print('y (km): ', y_max_velx)
    print('----------------------------------------')
```

# Problema 1. c)

```python
mean_velx = pd.DataFrame(np.zeros([n_rows, n_cols]), columns=column_indexes)
mean_vely = pd.DataFrame(np.zeros([n_rows, n_cols]), columns=column_indexes)

for file_index in file_indexes: 
    file_name_u = './release_gp_homework_data/OceanFlow/{}u.csv'.format(file_index)
    file_name_v = './release_gp_homework_data/OceanFlow/{}v.csv'.format(file_index)
    
    df_temp_u = pd.read_csv(file_name_u, names=column_indexes)
    df_temp_v = pd.read_csv(file_name_v, names=column_indexes)
    
    mean_velx.loc[:,:] = mean_velx.loc[:,:] + df_temp_u.loc[:,:] / 100
    mean_vely.loc[:,:] = mean_vely.loc[:,:] + df_temp_v.loc[:,:] / 100
```

```python
overall_mean_velx = mean_velx.mean().mean()
overall_mean_vely = mean_vely.mean().mean()
```

```python
print('Overall mean x component: ', overall_mean_velx)
print('Overall mean y component: ', overall_mean_vely)
```
